# Load 1240k and ISOGG Y SNPs.

```python
def load_1240k_y_df(path_snp = "/n/groups/reich/hringbauer/git/punic_aDNA/eigenstrat/punic0.v43.snp"):
    """Return 1240k SNP df"""
    df_snp0 = pd.read_csv(path_snp, delim_whitespace=True, header=None)
    df_snp0.columns = ["snp", "chr", "map", "pos", "ref", "alt"]
    df_1240k = df_snp0[df_snp0["chr"]==24]
    return df_1240k

def load_isogg_df(path_snp = "/n/groups/reich/hringbauer/git/y_chrom/data/eigenstrat/y_snps_all2020.snp"):
    df_isogg = pd.read_csv(path_snp, delim_whitespace=True, header=None)
    df_isogg.columns = ["snp", "chr", "map", "pos", "ref", "alt"]
    return df_isogg
```

```python
df_1240k = load_1240k_y_df()
print(f"ISOGG SNPs")
pos_min = np.min(df_1240k["pos"])
pos_max = np.max(df_1240k["pos"])
print(f"Accessible Y: {pos_min}-{pos_max}")
print(f"Length: {(pos_max - pos_min)/10**6:.4f} mB")
```

```python
df_isogg=load_isogg_df()
print(f"1240k SNPs")
pos_min = np.min(df_1240k["pos"])
pos_max = np.max(df_1240k["pos"])
print(f"Accessible Y: {pos_min}-{pos_max}")
print(f"Length: {(pos_max - pos_min)/10**6:.4f} mB")
```

# Make Figure of Coverage

```python
### Create the Binned Statistic
bins10 = np.arange(pos_min, pos_max, step=1e5)  # 0.5mB
counts = np.zeros(len(bins10), dtype="int")
```

```python
fs = 14

plt.figure(figsize=(12,6))
c1 = "blue"
c2 = "red"

ax = plt.gca()
ax.set_xlim([0, 3e7])
n, b = np.histogram(df_isogg["pos"], bins = bins10)
x = (b[1:] + b[:-1]) / 2

ax.plot(x, n,color=c1, linewidth=1.0)
ax.tick_params(axis='y', labelcolor=c1)
#ax1.set_ylim([-1, 800])
ax.set_ylabel("ISOGG SNPs per 100kb", color=c1, rotation=90, labelpad=fs*1.4, fontsize=fs)

ax1 = ax.twinx()  # instantiate a second axes that shares the same x-axis
n, b = np.histogram(df_1240k["pos"], bins = bins10)
x = (b[1:] + b[:-1]) / 2

ax1.plot(x, n,color=c2, linewidth=1.0)
ax1.tick_params(axis='y', labelcolor=c2)
#ax1.set_ylim([-1, 800])
ax1.set_ylabel("1240k SNPs per 100kb", color=c2, rotation=270, labelpad=fs*1.4, fontsize=fs)


ax.set_xlabel("Position Chromosome Y [bp]", fontsize=fs)
plt.show()
```

# Check how many ISOGG SNPs are near 1240k SNPs

```python
df_isogg=load_isogg_df()
df_1240k = load_1240k_y_df()
```

```python
pos_k=df_1240k["pos"].values
pos_i=df_isogg["pos"].values
```

```python
%%time
idx = np.searchsorted(pos_k, pos_i, side='right', sorter=None)
lower_dist = pos_i - pos_k[idx-1]
upper_dist =  pos_k[idx+1] - pos_i
both_dist = np.minimum(lower_dist,upper_dist)
```

```python
bins=np.linspace(-0.5,4999.5,401)
plt.figure()
plt.hist(both_dist, bins=bins, ec="k")
plt.show()
```

```python
distance = np.sort(both_dist)
quantile = np.arange(len(both_dist))/len(both_dist)
```

```python
plt.figure()
ax=plt.gca()
ax.plot(distance, quantile, lw=3)
ax.set_xlim([0,800])
ax.set_xlabel("Distance d to closest 1240k SNP [bp]")
ax.set_ylabel("Fraction ISOGG SNPs <d")
plt.show()
```

```python
delta = pos_k[1:]- pos_k[:-1]

bins=np.linspace(-0.5,4999.5,401)

plt.figure()
ax=plt.gca()
ax.hist(delta, bins=bins, ec="k")
ax.set_xlim([0,50])
plt.show()
```
